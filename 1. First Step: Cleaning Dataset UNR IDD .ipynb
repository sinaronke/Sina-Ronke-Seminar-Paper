```python
#Change all features (Switch ID, Port Number, Label, Binary Label) with string to integer to train machine learning models later
#Switch ID
switch_id_to_int = {"of:0000000000000001": 1, "of:0000000000000002": 2, "of:0000000000000003": 3, "of:0000000000000004": 4, "of:0000000000000005": 5, "of:0000000000000006": 6, "of:0000000000000007": 7, "of:0000000000000008": 8, "of:0000000000000009": 9, "of:000000000000000a": 10, "of:000000000000000b": 11, "of:000000000000000c": 12}
df["Switch ID"] = df["Switch ID"].map(switch_id_to_int)

#Port Number
port_number_to_int = {"Port#:1": 1, "Port#:2": 2, "Port#:3": 3,"Port#:4": 4}
df["Port Number"] = df["Port Number"].map(port_number_to_int)

#Label
label_to_int = {"Blackhole": 1, "Diversion": 2, "Normal": 3,"Overflow": 4, "PortScan": 5, "TCP-SYN": 6}
df["Label"] = df["Label"].map(label_to_int)

#Binary Label
binary_label_to_int = {"Attack": 1, "Normal": 2}
df["Binary Label"] = df["Binary Label"].map(binary_label_to_int)
```

```python
#Delete all columns with values "0" only
df = df.loc[:, (df != 0).any(axis=0)]
```

```python
#Delete all columns with negative values - some ML algorithms doesnt work with negative values
df=df[(df >= 0).all(1)]
```
